```python
#Assigning networkx to get a collection of nodes,and edges.
import networkx as nx

# Without import Matplotlib.pyplot, the graph will not display.
import matplotlib.pyplot as plt      

#Assigning a new graph by creating Picture variable.
Picture = nx.Graph()

#Creating a single color for every nodes on the map.
rainbowcolors = []

#Creating pink color to display the names clearly on the nodes.
rainbowcolors.append('pink')   

#Setting the correct positions with the names along with each nodes.
Picture.add_node('Linky', Photo = (2, 5)); Picture.add_node('LuLu', Photo = (1, 4));Picture.add_node('William', Photo = (3, 4));Picture.add_node('Aung', Photo = (2, 3));
Picture.add_node('Nima', Photo = (1, 2));Picture.add_node('Zhang', Photo = (3, 2));Picture.add_node('Joseph', Photo = (2, 1));Picture.add_node('MdUniq', Photo = (4, 3));
Picture.add_node('Chris', Photo = (5, 3));Picture.add_node('Grace', Photo = (6, 3))

#Assigning edges to attach all the given names and will be shown on the graph later.
Picture.add_edges_from([ ('Linky', 'LuLu'), ('Linky', 'William'), ('Linky', 'Joseph'), ('LuLu', 'William'), ('LuLu', 'Aung'), ('LuLu', 'Nima'),
('Nima', 'Aung'), ('Nima', 'Zhang'), ('Nima', 'Joseph'),('William', 'Aung'), ('William', 'MdUniq'), ('William', 'Zhang'), ('Zhang', 'Aung'), ('Zhang', 'MdUniq'), ('Zhang', 'Joseph'),
('Joseph', 'Aung'), ('Chris', 'MdUniq'), ('Chris', 'Grace')])

#Setting positions and attaches to all the given names with edges and displaying on the graphs.
Photo = nx.get_node_attributes(Picture, 'Photo')  

# If all the variable: picture, nodes, and edges are correct, and With-labels = True, this will print all the names
nx.draw(Picture,Photo, node_color = rainbowcolors, with_labels = True, edge_color='purple')
plt.show()
```
